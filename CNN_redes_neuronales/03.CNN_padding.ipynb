```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
```

```python
# Cargar imagen real en escala de grises
img = cv2.imread('cat.jpg', cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (128, 128))
```

```python
# Filtro detector de bordes verticales (Sobel)
kernel = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])
```

```python
# Convolución sin padding (valid)
convolved_valid = convolve2d(img, kernel, mode='valid')
```

```python
# Convolución con padding (same)
convolved_same = convolve2d(img, kernel, mode='same')
```

```python
# Visualización
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.title('Imagen Original')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Sin Padding (Valid)')
plt.imshow(convolved_valid, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Con Padding (Same)')
plt.imshow(convolved_same, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
from skimage.measure import block_reduce

# Leer imagen real
img = cv2.imread('cat.jpg', cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (128, 128))

# Filtro Sobel vertical
kernel = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])

# Convolución con padding SAME
convolved = convolve2d(img, kernel, mode='same')

# Aplicar MaxPooling 2x2
pooled = block_reduce(convolved, block_size=(2, 2), func=np.max)

# Visualización
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.title('Imagen original')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Después de convolución')
plt.imshow(convolved, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Después de MaxPooling')
plt.imshow(pooled, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
